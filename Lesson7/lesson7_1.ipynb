```python
import requests
from requests import Response
from requests.exceptions import RequestException,HTTPError # 下載失敗時的例外處理
url = 'https://data.ntpc.gov.tw/api/datasets/010e5b15-3823-4b20-b401-b1cf000550c5/csv?page=0&size=1000'

try:
    r:Response = requests.request('GET',url)
    r.raise_for_status()   
except HTTPError as e:
    print(e)
except RequestException as e:
    print(e)
else:
    print("下載成功")
  
```

#### 回傳下載成功失敗

```python
from requests import Response
from requests.exceptions import RequestException,HTTPError # 下載失敗時的例外處理
try:
    r_1:Response = requests.request('Get',url)
    r_1.raise_for_status()
except HTTPError as e:
    print(e)
except RequestException as e:
    print(e)
else:
    print("DownLoad Successful")
```

#### CSV轉換成python 資料結構轉換
    1. IO 套件 io.stringIO
    2. CSV檔案 csv.DictReader
    3. for in loop 抓取Dict內的資料轉換成list

```python
from io import StringIO
file = StringIO(r_1.text)
file 
```

```python
from csv import DictReader
reader = DictReader(file)
list_reader = reader 
```

```python
for row in list_reader: # reader是一個generator的實體
    print(row)
```
