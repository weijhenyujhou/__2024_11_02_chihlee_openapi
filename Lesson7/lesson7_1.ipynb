### APi request 程式說明

```python
import requests
url = 'https://data.ntpc.gov.tw/api/datasets/34b402a8-53d9-483d-9406-24a682c2d6dc/csv?page=0&size=1000'
r = requests.get(url)
r.text
print(r.text)
```

#### 回傳下載成功失敗

```python
from requests import Response
r_1:Response = requests.request('Get',url)
if r_1.status_code == 200:
    print("DownLoad Successful")
else:
    print("DownLoad Failed")
```

#### CSV轉換成python 資料結構轉換
    1. IO 套件 io.stringIO
    2. CSV檔案 csv.DictReader
    3. for in loop 抓取Dict內的資料轉換成list

```python
from io import StringIO
file = StringIO(r_1.text)
file 
```

```python
from csv import DictReader
reader = DictReader(file)
list_reader = reader 
```

```python
for row in list_reader: # reader是一個generator的實體
    print(row)
```
